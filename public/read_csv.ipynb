```python
df = pd.read_csv('product_list_csv.csv')
```

```python
# กำหนดบาร์โค้ดที่ต้องการค้นหา
search_code = "8858705604595"

# ค้นหาจากทั้ง 3 คอลัมน์บาร์โค้ด
result = df[
    (df['Bar Code EA'] == search_code) |
    (df['Bar Code DSP'] == search_code) |
    (df['Bar Code CS'] == search_code)
]

# แสดงผลลัพธ์
print(result)
```
